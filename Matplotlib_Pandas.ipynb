**Matplotlib**

*   vẽ biểu đồ, đồ thị và hình ảnh trong khoa dữ liệu




```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1,2,3,4,5])
y = np.array([4,9,6,3,8])
y2 = np.array([8,5,3,6,9])
```

```python
plt.plot(x,y)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,c="r")
plt.title("Biểu đồ doanh thu")
plt.xlabel("ABC")
plt.ylabel("XYZ")
plt.subplot(1,2,2)
plt.plot(x,y2)
```

```python
#x = np.array([1,2,3,4,5])
#y2 = np.array([8,5,3,6,9])
```

```python
plt.plot(x,y,x,y2)
```

**Bar**

```python
plt.bar(x,y)
```

```python
plt.bar(x,y)
plt.title("Biểu đồ doanh thu")
plt.xlabel("ABC")
plt.ylabel("XYZ")
plt.legend(["Giá nhà"])
```
